## System Model 4 A/B
- Path graph: $|\lambda_{\max}(A)|=0.8$
- No actuators
- Cycle (A), Biased-Cycle (B) state-MPL noise
- Initial state distribution - identity covariance

[Model List](#Model-list)

```python
nx = 8

cycle = create_graph(nx, type='cycle')
path = create_graph(nx, type='path')

alphai = [0.1]

X0 = 5*np.identity(nx)


S = system_package(A_in=0.8*path['A'], alphai_in=alphai, Ai_in=cycle['Adj'], X0_in=X0, label_in='System Model 4A')
sys_to_file(S)

S = system_package(A_in=0.8*path['A'], alphai_in=alphai, Ai_in=cycle['Adj'] + np.diag(np.linspace(1, 0, num=nx)), X0_in=X0, label_in='System Model 4B')
sys_to_file(S)
```
